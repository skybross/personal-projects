Below is brief description of each of the columns:

- **Initials**: The doctor's first and last initials.
- **Satisfaction**: The average patient satisfaction rating, from 1 (low) to 5 (high). If (and only if) no patients have reviewed this doctor, the value is 0.
- **Num Ratings**: The number of satisfaction ratings submitted for this doctor, from which the aggregate rating is calculated.
- **Specialty**: The doctor's primary field of medical practice.
- **Experience**: The number of years for which the doctor has been practicing.
- **Medicare?**: Describes whether the doctor accepts Medicare.
- **Medicaid?**: Describes whether the doctor accepts Medicaid.
- **Location**: The city where the doctor's office is located.
- **Bus?**: Describes whether the office is easily accessible by public buses.
- **Spanish?**: Describes whether the doctor speaks Spanish (in addition to English).
- **Med School**: The medical school from which the doctor graduated.
- **Num Awards**: The number of awards this doctor has received.
- **Gender**: The doctor's gender.
